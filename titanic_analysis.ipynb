```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

### 1. Carregando a base TITANIC

```python
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df_titanic = pd.read_csv(url)
```

### 2. Filtrando apenas 'Age' e 'Fare'

```python

df = df_titanic[['Age', 'Fare']].copy()
```

### 3. Removendo linhas com Idade (Age) faltando (NaN)

```python
df = df.dropna(subset=['Age'])
```

### 4. Definindo X (idade) e y (preço da passagem)

```python

X = df[['Age']] 
y = df['Fare']
```

### 5. Treinando a Regressão Linear

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

### 6. Treinando a Árvore Full

```python
tree_full = DecisionTreeRegressor(random_state=42)
tree_full.fit(X, y)
```

### 7. Treinando a Árvore

```python
tree_depth3 = DecisionTreeRegressor(max_depth=3, random_state=42)
tree_depth3.fit(X, y)
```

### Gráfico para visualização e comparação

```python
# Criando um array de idades (do mínimo ao máximo) para traçar as linhas de predição

import numpy as np
X_test = np.arange(df['Age'].min(), df['Age'].max() + 1)[:, np.newaxis]

plt.figure(figsize=(10, 6))

# Plotando os dados reais (dispersão)
plt.scatter(X, y, color='lightgray', label='Dados reais (Idade vs Preço)', alpha=0.6)
# Plotando as predições de cada modelo
plt.plot(X_test, lin_reg.predict(X_test), color='blue', label='Regressão Linear', linewidth=2)
plt.plot(X_test, tree_full.predict(X_test), color='red', label='Árvore Full (Overfitting)', alpha=0.5)
plt.plot(X_test, tree_depth3.predict(X_test), color='green', label='Árvore Depth=3', linewidth=2)

plt.title('Comparação de Modelos: Preço da Passagem (Fare) baseado na Idade (Age)')
plt.xlabel('Idade (Age)')
plt.ylabel('Preço da Passagem (Fare)')
plt.legend()
plt.tight_layout()
```
